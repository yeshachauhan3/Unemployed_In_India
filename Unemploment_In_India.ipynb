#Heatmap

#Histogram According to the Area

```python
data.columns=['Region', 'Date','Frequency' ,'Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate', 'Area']

plt.title('Indian Unemployment')


sns.histplot(x ='Estimated Employed', hue='Area', data= data)

plt.show()
```

#Histogram accordinh to the Region

```python
data.columns=['Region', 'Date','Frequency' ,'Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate', 'Area']

plt.title('Indian Unemployment')

plt.figure(figsize=(9, 14))
sns.histplot(x ='Estimated Employed', hue='Region', data= data)

plt.show()
```

#Bar graph on Date and Estimated  Employee

```python
data['Date'] = pd.to_datetime(data['Date'])


data_grouped = data.groupby('Date')['Estimated Employed'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Date', y='Estimated Employed', data=data_grouped, palette='viridis')
plt.title('Total Employed by Date')
plt.xlabel('Date')
plt.ylabel('Total Employed')
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees for better readability
plt.tight_layout()
plt.show()
```
